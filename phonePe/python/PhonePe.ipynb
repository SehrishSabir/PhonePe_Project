```python
service_revenue = all_transaction.groupby('Service Type')['Amount'].sum().sort_values(ascending=False)

service_revenue.plot(kind='bar')
plt.title('Revenue by Service')
plt.ylabel('Total Amount')
plt.show()
```

```python
adoption = all_transaction.groupby('Service Type')['User_ID'].nunique()

adoption.plot(kind='bar')
plt.title('Service Adoption Rate')
plt.ylabel('Unique Users')
plt.show()
```

```python
all_transaction['Payment_Status'].value_counts().plot(kind='pie', autopct='%2.1f%%')

plt.title('Payment Status Distribution')
plt.ylabel('')
plt.show()
```

```python
all_transaction['Date'] = pd.to_datetime(all_transaction['Date'])

monthly = all_transaction.groupby(
    all_transaction['Date'].dt.to_period('M')
)['Amount'].sum()

monthly.plot(kind='line', marker='o')
plt.title('Monthly Transaction Trend')
plt.show()
plt.show()
```

```python
top_users = all_transaction.groupby('User_ID')['Amount'].sum()\
                            .sort_values(ascending=False)\
                            .head(10)

top_users.plot(kind='bar')
plt.title('Top 10 Customers')
plt.show()
```

```python
all_transaction['Reason'].value_counts().head(10).plot(kind='bar')

plt.title('Top Transaction Reasons')
plt.show()
```

```python
df = pd.merge(df_users, all_transaction, on='User_ID')

sns.boxplot(data=df,
            x=pd.cut(df['Age'],
                     bins=[18,25,35,45,60]),
            y='Amount')

plt.title('Age Group vs Transaction Amount')
plt.show()
```
